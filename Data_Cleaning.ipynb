```python
data = pd.read_csv("/content/drive/MyDrive/Dataset/Data Mining/Chapter03DataSet.csv")
```

```python
# Melihat jumlah data yang hilang pada setiap atribut
data.isnull().sum()
```

```python
# Mengisi data yang kosong
data['Read_News'].fillna("Y", inplace=True)
data['Online_Shopping'].fillna("N", inplace=True)
data['Online_Gaming'].fillna("Y", inplace=True)

# Mengubah kolom numerik menjadi nominal
data['Twitter'] = data['Twitter'].replace('99', 'N')

# Menghapus kolom 'Other_Social_Network'
data.drop(columns=['Other_Social_Network'], inplace=True)
```
